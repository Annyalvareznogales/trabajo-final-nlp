## Tarea de clasificación en inglés

Para el caso de la tarea de clasificación en inglés, las ideas que hemos desarrollado en nuestro proyecto consisten en:
- **Preprocesamiento de Datos:** Se limpiaron los conjuntos de entrenamiento y prueba normalizando los textos a minúsculas, o eliminando elementos como URLs, saltos de línea, múltiples espacios en blanco, caracteres especiales y números (excepto comas y puntos). Esto facilitó notablemente la interpretación correcta de los textos, asegurando su consistencia.

- **Manejo del desequilibrio de Clases:** Se identificó un desequilibrio significativo en los datos, con una clara subrepresentación de la clase CONSPIRANCY. Para abordar este problema, se implementó una estrategia de data augmentation, pero con un enfoque diferenciador: los textos originales fueron traducidos a otro idioma utilizando modelos preentrenados de Hugging Face y luego traducidos de vuelta al idioma original. Esta técnica producía pequeñas variaciones que enriquecen nuestros datos, generando datos similares pero no idénticos para mitigar riesgos de overfitting. En concreto, se ha decidido traducir los textos en inglés al alemán, y los textos en español al inglés.  

- **Adición de Información Contextual:** Dada la subjetividad de la tarea, se exploraron estrategias para enriquecer los datos con información adicional. Inicialmente, se consideró utilizar modelos preentrenados para extraer características textuales como emociones, pero esta opción se descartó debido a su dependencia del sesgo del modelo seleccionado. Optamos por incorporar léxicos en inglés (emocionales y morales) que relacionan palabras clave en los textos con emociones y valores morales. La información extraída de estos léxicos se incluyó como un prompt añadido al final del texto original, indicando qué emoción y valor moral reflejaba el contenido.  

- **Experimentos y Ajustes:** Se realizaron múltiples experimentos para identificar el modelo preentrenado más efectivo (comparando entre BERT-base-uncased y RoBERTa), evaluar el impacto del aumento de datos y ajustar hiperparámetros clave. También se exploraron configuraciones avanzadas, como el aumento del tamaño de batch, para mejorar los resultados del modelo.  

- **Cross-Validation y Estrategias de Validación.** Para garantizar una evaluación robusta del modelo y evitar problemas de overfitting, implementamos diferentes técnicas de Cross-Validation. Inicialmente, utilizamos K-Fold Cross-Validation, dividiendo el conjunto de datos en k conjuntos para entrenar y validar el modelo en múltiples iteraciones. Este enfoque permitió evaluar el rendimiento del modelo de manera uniforme en diferentes particiones de los datos. Sin embargo, al observar la distribución desequilibrada de las clases en el dataset (particularmente la subrepresentación de la clase CONSPIRANCY), decidimos emplear una variante más avanzada: Stratified K-Fold Cross-Validation (Sk-Fold). Esta técnica garantiza que la proporción de las clases en cada pliegue sea similar a la del conjunto de datos original, preservando la representatividad de las clases en las particiones de entrenamiento y validación. La incorporación de Sk-Fold fue crucial para mejorar la estabilidad de los resultados, especialmente en métricas como el F1-Score y el MCC, ya que permitió al modelo aprender patrones más consistentes incluso en clases minoritarias.

- **Resultados y métricas.** Los resultados fueron evaluados utilizando métricas como F1-score, y el coeficiente MCC, según lo solicitado en la competencia original. El mejor desempeño se obtuvo con el modelo RoBERTa, en combinación con el aumento de datos mediante traducciones y la adición del contexto emocional y moral extraído de los léxicos. Este enfoque superó el baseline de la competencia PAN.  

- **Experimentación Adicional.** Como parte de la exploración de técnicas avanzadas, modificamos la estructura base de RoBERTa añadiendo capas de atención múltiple y convoluciones, (siguiendo un enfoque basado en el paper Cheruku, R. et al. (2023) Sentiment classification with modified Roberta and recurrent neural networks - multimedia tools and applications, SpringerLink. Available at: https://link.springer.com/article/10.1007/s11042-023-16833-5 (Accessed: 22 November 2024) . Aunque no se lograron mejoras significativas, este experimento permitió aprender cómo alterar arquitecturas para tareas con alta subjetividad, como clasificar emociones y añadir metadatos textuales. También se realizó un experimento combinando las representaciones [CLS] de BERT y RoBERTa. Los embeddings de ambas arquitecturas (768 dimensiones cada uno) fueron concatenados para formar un vector de 1536 dimensiones, que se introdujo en una capa totalmente conectada para la clasificación binaria

Seguidamente, una vez que tengo los datos correctamente estructurados, vamos a proceder a hacer diferentes comparaciones para apreciar las mejoras de  
los resultados con distintos parámetros y cambios. Para ello, graficaremos estas funciones indicando el baseline propuesto en la competición.

```python
baselines = {'MCC': 0.7964, 'F1-Score (Macro Avg)': 0.8975, 'F1-Score (CONSPIRANCY)': 0.8632, 'F1-Score (CRITICAL)': 0.9318}
```

```python
def graficar_basico(title, models_to_compare, metrics):
    filtered_df = df[df['Model'].isin(models_to_compare)]
    filtered_melted = filtered_df.melt(id_vars=["Model"], value_vars=metrics, var_name="Metric", value_name="Value")
    plt.figure(figsize=(12, 5))
    sns.barplot(data=filtered_melted, x="Metric", y="Value")
    
    # Mostrar baseline para MCC y F1-Score
    palette = sns.color_palette('Paired', len(metrics))
    for i, metric in enumerate(metrics):
        if metric in baselines:
            plt.axhline(y=baselines[metric], color=palette[i], linestyle='--', label=f'Baseline {metric}')
    plt.legend(bbox_to_anchor=(1.05, 1))
    
    plt.title(title)
    plt.ylabel("Valor")
    plt.xlabel("Métrica")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

## Bert con datos iniciales: sin contexto ni data augmentation

En primer lugar, ofreceremos una visión general de las métricas evaluadas, comparando nuestros datos iniciales proporcionados con el modelo BERT. Las líneas de los baselines sirven como referencia para evaluar la efectividad del modelo en comparación con las expectativas o los valores proporcionados para cada métrica.

```python
metrics = ["MCC", "Accuracy", "Precision (CONSPIRANCY)", "Recall (CONSPIRANCY)", "F1-Score (CONSPIRANCY)", "Precision (CRITICAL)", "Recall (CRITICAL)",\
           "F1-Score (CRITICAL)", "Precision (Macro Avg)", "Recall (Macro Avg)", "F1-Score (Macro Avg)", "Precision (Weighted Avg)",\
           "Recall (Weighted Avg)", "F1-Score (Weighted Avg)"]
models_to_compare = ['test_evaluation_results_bert_nocontext_nodataaugmentation']
title=f"Comparación de Métricas de Bert caso inicial"

graficar_basico(title,models_to_compare, metrics)
```

```python
baselines_en = {'MCC': 0.7964, 'F1-Score (Macro Avg)': 0.8975, 'F1-Score (CONSPIRANCY)': 0.8632, 'F1-Score (CRITICAL)': 0.9318}
```

```python
def plot_metrics_comparison(df, models, metrics, baselines=baselines_en):
    colors = {}
    df_filtered = df[df['Model'].isin(models)]

    #Calculo dinamico de filas
    ncols = 4
    nrows = (len(metrics) + ncols - 1) // ncols
    
    fig, axes = plt.subplots(nrows, ncols, figsize=(5 * ncols, 5*nrows))
    axes = axes.flatten()

    if len(metrics) == 1:
        axes = [axes]
    
    for i, metric in enumerate(metrics):
        ax = axes[i]
        
        for model in models:
            model_data = df_filtered[df_filtered['Model'] == model]
            line, = ax.plot(model_data['Model'], model_data[metric], label=model, marker='o', markersize= 8)
            colors[model] = line.get_color()

            # Añadir etiqueta con el valor decimal en cada punto
            for x, y in zip(model_data['Model'], model_data[metric]):
                ax.text(x, y, f'{y:.6f}', color='black', ha='center', va='bottom', fontsize=12)
        
        ax.set_title(f'{metric} Comparison', fontsize=16)
        ax.set_ylabel(metric, fontsize=14)
        ax.set_xticks([])
        ax.tick_params(axis='y', labelsize=14)

        # Calcular los valores mínimo y máximo de la métrica
        metric_min = min(df_filtered[metric].min(), baselines[metric]) if metric in baselines else df_filtered[metric].min()
        metric_max = max(df_filtered[metric].max(), baselines[metric]) if metric in baselines else df_filtered[metric].max()

        # Extender los límites en un 20% por encima y por debajo de los valores mínimo y máximo
        margin = 0.2 * (metric_max - metric_min)
        ax.set_ylim(metric_min - margin, metric_max + margin)

        #Añadir la linea para indicar el baseline
        if metric in baselines:
            ax.axhline(y=baselines[metric], color='red', linestyle='--', label=f'Baseline {metric}')
            ax.legend(loc='lower right', fontsize=10, labels=[f'Baseline = {baselines[metric]:.4f}'])

        #Añadir leyenda comun a todas las tablas
        handles = []
        for model in models:
            line = Line2D([0], [0], color=colors[model], lw=2, label=model)
            handles.append(line)
        fig.legend(handles=handles, loc='center', ncol=1, bbox_to_anchor=(0.5, -0.15), fontsize=20)

    # Ocultar los ejes que no se usan (gráficos vacíos)
    for j in range(i + 1, len(axes)):  # Desde el siguiente gráfico hasta el final
        axes[j].set_visible(False)

    plt.tight_layout()
    plt.show()
```

## BERT con Cross-Validation

### K-fold Cross-Validation + información emociones

En primer lugar, hemos decidido enriquecer nuestro conjunto de datos incorporando información adicional sobre las emociones reflejadas en el texto, tal como se explicó anteriormente. Esta adición tiene como objetivo mejorar la capacidad del modelo para identificar y clasificar emociones, lo que podría ofrecer una comprensión más profunda del contexto emocional en el que se expresan las ideas. Además, hemos implementado el concepto de k-fold cross-validation con el objetivo de optimizar la evaluación de nuestro modelo y garantizar que los resultados sean más robustos y generalizables. De esta manera, conseguimos una evaluación más fiable del rendimiento del modelo, evitando el overfitting y mejorando su capacidad para generalizar a nuevos datos.

Además, hemos decidido modificar el valor de la longitud máxima de los textos (max_len) del modelo, así como su tasa de aprendizaje (learning rate), con el objetivo de optimizar el rendimiento del modelo ajustando los hiperparámetros clave. Al aumentar el max_len, buscamos mejorar la capacidad del modelo para manejar secuencias de longitud variable, capturando mayor contexto, a pesar de la posibilidad de overfitting que esto genera (debido a que es compensado mediante cross-validation). Por otro lado, al ajustar la tasa de aprendizaje, intentamos encontrar el balance adecuado entre la rapidez de convergencia y la estabilidad del entrenamiento, lo que podría resultar en una mayor precisión y una convergencia más eficiente. En cualquiera de los casos, vemos que existe una notable mejoría respecto a nuestros datos iniciales, y que además para `max_len=512` y `learning_rate=2e5` tiene un desempeño muy cercano a los baselines.

```python
metrics = ['MCC', 'Accuracy', 'F1-Score (Macro Avg)', 'F1-Score (CONSPIRANCY)', 'F1-Score (CRITICAL)']
models_to_compare = ['test_evaluation_results_bert_nocontext_nodataaugmentation',
                     'test_evaluation_results_256_2e5_emotions_bert_kfold_nodataaugmentation',
                    'test_evaluation_results_512_1e5_emotions_bert_kfold_nodataaugmentation',
                    'test_evaluation_results_512_2e5_emotions_bert_kfold_nodataaugmentation']
plot_metrics_comparison(df, models_to_compare, metrics)
```

A pesar de ello, seguimos observando una diferencia entre el f1-score para CRITICAL y CONSPIRANCY, es por ello que vamos a incorporar nuestra técnica de data augmentation con el propósito de balancear los datos. Mantendremos el `max_len=512` y `learning_rate=2e5` ya que son los que mejores resultados nos han dado.

```python
metrics = ['MCC', 'F1-Score (Macro Avg)', "F1-Score (CONSPIRANCY)", "F1-Score (CRITICAL)"]
models_to_compare = ['test_evaluation_results_256_2e5_emotions_bert_kfold_nodataaugmentation',
                    'test_evaluation_results_512_1e5_emotions_bert_kfold_nodataaugmentation',
                     'test_evaluation_results_512_2e5_emotions_bert_kfold_nodataaugmentation',
                     'test_evaluation_results_512_2e5_emotions_morals_bert_kfold_dataaugmentation']

plot_metrics_comparison(df, models_to_compare, metrics)
```

Observamos que no existe un gran aporte en cuanto a los resultados, ya que de hecho no produce ninguna mejora. Sin embargo, sigue resultando un poco mejor que las opciones con un menor learning rate y max_len. 

### Sk-fold Cross-Validation + información emociones + data augmentation

Como última prueba para nuestro modelo BERT, introduciremos el concepto de Stratified K-fold Cross-Validation (skfold), con el propósito de obtener una evaluación más precisa y equilibrada del rendimiento del modelo. Este enfoque garantiza que cada pliegue (fold) contenga una distribución proporcional de las clases, lo que resulta especialmente beneficioso en conjuntos de datos desbalanceados, donde algunas clases pueden ser menos representadas que otras. Al usar Stratified K-fold, minimizamos el sesgo de la muestra y mejoramos la generalización del modelo, asegurando que el modelo no se vea influenciado negativamente por la distribución desproporcionada de clases.

```python
metrics = ['MCC', 'F1-Score (Macro Avg)', 'F1-Score (CONSPIRANCY)', 'F1-Score (CRITICAL)']
models_to_compare = ['test_evaluation_results_256_2e5_emotions_bert_kfold_nodataaugmentation',
                    'test_evaluation_results_512_1e5_emotions_bert_kfold_nodataaugmentation',
                    'test_evaluation_results_512_2e5_emotions_bert_kfold_nodataaugmentation',
                    'test_evaluation_results_512_2e5_emotions_morals_bert_skfol_dataaugmentation']
plot_metrics_comparison(df, models_to_compare, metrics)
```

Podemos apreciar una pequeña mejora para la clasificación de la etiqueta CONSPIRANCY, sin embargo no resulta relevante para nuestro experimento, debido a que en líneas generales los resultados podríamos considerar que son muy similares o prácticamente iguales.

Como parte de las líneas futuras, y con acceso a mayores recursos computacionales, sería interesante explorar modelos más robustos, como BERT-base-multilingual-cased, para el cual ya hemos preparado el código necesario.
